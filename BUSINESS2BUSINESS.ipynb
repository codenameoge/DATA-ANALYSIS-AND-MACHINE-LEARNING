HANDLING THE MISSING VALUES FOR THE CATEGORICAL VALUES

```python
data.order_item_id= data.order_item_id.fillna(data.order_item_id.mean())
data.price =data.price.fillna(data.price.mean())
data.payment_value=data.payment_value.fillna(data.payment_value.mean())
```

HANDLING MISSING VALUES IN NUMERICAL VALUES

```python
data['product_id'].fillna(data['product_id'].mode()[0],inplace=True)
data['product_category_name_english'].fillna(data['product_category_name_english'].mode()[0],inplace=True)
data['seller_id'].fillna(data['seller_id'].mode()[0],inplace=True)
data['payment_type'].fillna(data['payment_type'].mode()[0],inplace=True)
data['order_delivered_customer_date'].fillna(data['order_delivered_customer_date'].mode()[0], inplace=True)
```

USING CAPPING AND FLOORING METHOD TO HANLE THE OUTLIERS

```python
np.percentile(data.price,[99])
```

```python
np.percentile(data.price,[99])[0]
```

```python
uv = np.percentile(data.price,[99])[0]
```

```python
data[(data.price>uv)]
```

```python
data.price[(data.price> 3*uv)] = 3*uv
```

```python
np.percentile(data.payment_value,[1])[0]
```

```python
lv = np.percentile(data.payment_value,[1])[0]
```

```python
data[(data.payment_value< lv)]
```

```python
data.payment_value[(data.payment_value <0.3*lv)] = 0.3*lv
```
